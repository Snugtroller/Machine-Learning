```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import keras
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import LeakyReLU,PReLU,ELU
from keras.layers import Dropout
from sklearn.metrics import confusion_matrix,accuracy_score
```

```python
data=pd.read_csv("Churn_Modelling.csv")
data.head(5)
```

```python
X=data.iloc[:,3:13]
X.head(5)
```

```python
y=data.iloc[:,13]
y.head(5)
```

For converting in True or False.

```python
geography=pd.get_dummies(X["Geography"],drop_first=True)
gender=pd.get_dummies(X["Gender"],drop_first=True)
```

```python
X=pd.concat([X,geography,gender],axis=1)
X.head(5)
```

```python
X=X.drop(["Geography","Gender"],axis=1)
X.head(5)
```

```python
sc=StandardScaler()
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=0)
X_train=sc.fit_transform(X_train)
X_test=sc.transform(X_test)
```

```python
X_train.shape
```

```python
fn = Sequential()
fn.add(Dense(units=8, kernel_initializer="he_uniform", activation='relu', input_dim=11))
fn.add(Dense(units=8, kernel_initializer="he_uniform", activation='relu'))
fn.add(Dense(units=1, kernel_initializer='glorot_uniform', activation='sigmoid'))
fn.compile(optimizer="adam",loss=keras.losses.binary_crossentropy,metrics=['accuracy'])
fn.fit(X_train,y_train,batch_size=10,epochs=100)
```

```python
y_pred=fn.predict(X_test)
y_pred=(y_pred>0.5)
```

```python
confusion_matrix(y_test,y_pred)
```

```python
  accuracy_score(y_test,y_pred)
```
