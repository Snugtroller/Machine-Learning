```python
import pandas as pd
import numpy as np
from tensorflow import keras
import matplotlib.pyplot as plt
```

Imported the dataset from keras

```python
(X_train, y_train) , (X_test, y_test) = keras.datasets.mnist.load_data()
```

This particular image has a size of 28x28

```python
X_train[0]
```

We can conclude by saying that there are 60000 images having size of 28x28

```python
X_train.shape
```

```python
X_test.shape
```

The X_train and X_test are in 3-D array so we will be converting to a 1-D for avoiding complexity

```python
X_train_flatten=X_train.reshape(60000,28*28)
X_test_flatten=X_test.reshape(10000,28*28)
```

```python
X_train_flatten[0]
```

We are normalizing the arrays into decimals i.e.., numbers between 0's and 1's

```python
X_train_normal=X_train_flatten/255
X_test_normal=X_test_flatten/255
```

```python
X_train_normal[0]
```

Using keras to fit the modal in the neural network, we will be taking 10 as output as there are numbers between 0 to 9. The input layer will be having all the flatten data i.e.., 28*28 data.

```python
model=keras.Sequential([keras.layers.Dense(10,input_shape=(28*28,),activation="sigmoid")])
model.compile(optimizer="adam",loss="sparse_categorical_crossentropy",metrics=['accuracy'])
model.fit(X_train_normal,y_train,epochs=10)
```

```python
y_pred=model.predict(X_test_normal)
```

```python
np.argmax(y_pred[1])
```

Hence the y_pred is matching the X_test data which will be a number 7. We use argmax to get the highest probability from our y_pred.

```python
X_test[1]
```

```python
model.evaluate(X_test_normal,y_test)
```
